$$
\begin{bmatrix}
Var(x_1) & Cov(x_1, x_2) & \dotsb & Cov(x_1, x_n)\\
Cov(x_1, x_2) & Var(x_2) & \dotsb & Cov(x_2, x_n)\\
\vdots & \vdots & \ddots & \vdots\\
Cov(x_n, x_1) & Cov(x_n, x_2) & \dotsb & Var(x_n)\\
\end{bmatrix}
$$


$$
Cov(X_1, X_2)= \frac{\sum_{i=0}^{n} (X_1 - \bar{X_1})(X_2 - \bar{X_2})}{n}
$$

$$
Var(X_1)= \frac{\sum_{i=0}^{n} (X_1 - \bar{X})^2}{n}
$$

##5. Training Logistic Regression on MNIST

````python
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import fetch_openml

mnist = fetch_openml('mnist_784')
X = mnist.data
y = mnist.target.astype(np.int)
X_filtered = X[(y == 0) | (y == 8)]
y_filtered = y[(y == 0) | (y == 8)]

split = 6000  # Usamos las primeras 6000 imágenes para entrenamiento
X_train, X_test = X_filtered[:split], X_filtered[split:]
y_train, y_test = y_filtered[:split], y_filtered[split:]

model = LogisticRegression(random_state=0)
model.fit(X_train, y_train)


print("_______________________")
accuracy = model.score(X_test, y_test)
print("_______________________")
print(f"Accuracy: {accuracy}")
````

````python
from keras.datasets import mnist
import numpy as np
from sklearn.linear_model import LogisticRegression

(x_train, y_train), (x_test, y_test) = mnist.load_data()

# seleccionar solo las imágenes y etiquetas de los dígitos 0 y 8
idx = np.concatenate((np.where(y_train == 0)[0], np.where(y_train == 8)[0]))
x_train = x_train[idx]
y_train = y_train[idx]

# aplanar las imágenes y estandarizar los datos
x_train = x_train.reshape(x_train.shape[0], -1)
x_train = x_train.astype('float32') / 255.

# asignar etiquetas binarias para 0 y 8
y_train_binary = np.zeros(y_train.shape)
y_train_binary[y_train == 8] = 1

# entrenar una regresión logística
logreg = LogisticRegression(max_iter=1000)
logreg.fit(x_train, y_train_binary)

# seleccionar solo las imágenes y etiquetas de los dígitos 0 y 8
idx = np.concatenate((np.where(y_test == 0)[0], np.where(y_test == 8)[0]))
x_test = x_test[idx]
y_test = y_test[idx]

# aplanar las imágenes y estandarizar los datos
x_test = x_test.reshape(x_test.shape[0], -1)
x_test = x_test.astype('float32') / 255.

# asignar etiquetas binarias para 0 y 8
y_test_binary = np.zeros(y_test.shape)
y_test_binary[y_test == 8] = 1

# evaluar el modelo en el conjunto de datos de prueba
score = logreg.score(x_test, y_test_binary)
print('Logistic Regression Accuracy: {:.2f}'.format(score))
````

It is a central problem in statistics, but it is sensitive to sparse errors with large magnitudes. 

Robust PCA is a technique that can effectively separate the background from the foreground by decomposing an image into a low-rank matrix (which contains the background) and a sparse matrix (which contains the foreground and noise). This approach can lead to more accurate and robust feature extraction and analysis. Principal Component Pursuit (PCP), solves the robust PCA problem.

The background in images can be highly variable or contain noise, making it difficult to isolate the foreground object or accurately extract features. ##10. What are the underlying mathematical principles behind LDA? What is it useful for?


Firstable, it is important to distinguis between Linear Discriminant Analysis which is a predictive modeling algorithm for multi-class classification. This one can be also used in dimensionality reduction providing a projection of a training dataset that best separates the examples by their assigned class.

The math behind LDA is that the model seeks to find a linear combination of input variables that achieves the maximum separation for samples between classes (class centroids or means) and the minimum separation of samples within each class.

What I personally find really fascinating here is the fact that we would have the linear combination (because we are refering to a 2 dimensional plane) that best separates the classes using the Eigenvector and Eigenvalues. 

Example: 

```
Red Circles:
(2, 3)
(3, 4)
(4, 5)
(5, 6)

Blue Squares:
(1, 1)
(2, 2)
(3, 3)
(4, 4)

```

After having a toy dataset which contains two different types of circles: blue and red. We want to find the linear condition that satisfies the best separation between the ones:

```
Red Circles Mean: (3.5, 4.5)
Blue Squares Mean: (2.5, 2.5)
```

Then measuring the differences between the means of the two classes to calculate the between-class scatter matrix

```
SB = (Red Mean - Blue Mean) * (Red Mean - Blue Mean)T
   = (1, 2) * (1, 2)T
   = [1 2]   [1]
     [2 4] * [2]
             [4]
   = [1 2]
     [2 4]

```


calculate the within-class scatter matrix, which measures the variation within each class:

```
#SW = Σ(Red Data - Red Mean) * Σ(Red Data - Red Mean)T 
       + Σ(Blue Data - Blue Mean) * Σ(Blue Data - Blue Mean)T
   = (2, 3) * (2, 3)T + (3, 4) * (3, 4)T + (4, 5) * (4, 5)T + (5, 6) * (5, 6)T
       + (1, 1) * (1, 1)T + (2, 2) * (2, 2)T + (3, 3) * (3, 3)T + (4, 4) * (4, 4)T
   = [ 2  3]   [ 2  3]T   [ 1  1]   [ 1  1]T
     [ 3  4] * [ 3  4]T + [ 2  2] * [ 2  2]T
     [ 4  5]   [ 4  5]T   [ 3  3]   [ 3  3]T
     [ 5  6]   [ 5  6]T   [ 4  4]   [ 4  4]T
   = [ 4   4]
     [ 4  4]

```


And then calculate the SW^-1 * SB to obtain the result:

```
SW^-1 * SB = [1/4  1/4] * [1 2] = [3/4 3/4]
             [1/4  1/4]   [2 4]   [3/4 3/4]

Eigenvalues: λ1 = λ2 = 1.5
Eigenvectors: v1 = (1, 1), v
```
